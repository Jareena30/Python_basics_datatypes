```python
#while loop
c=5
while c<10 :
  print(c)
  c=c+1
```

```python
#break keyword
c=0
while c<10 :
  print(c)
  if c==4:
    break
  c=c+1
```

```python
#continue keyword
c=0
while c<10 :
    if c==4:
     continue
    print(c)
    c=c+1
```

```python
#pass keyword
c=0
while c<10 :
    if c==4:
     pass
    print(c)
    c=c+1
```
